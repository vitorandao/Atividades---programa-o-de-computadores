**19) Troco ou valor insuficiente.** Leia o preço de um produto e o valor pago. Se o pagamento for
suficiente, exiba o troco; caso contrário, informe quanto falta.

```python
preço=float(input("digite o preço do produto:"))
pago=float(input("digite o valor pago:"))
if  pago>=preço:
    print("troco:",pago-preço)
else:
    print("falta:",preço-pago)
```

**20) Triângulo válido.** Leia três valores representando lados de um triângulo. Informe se eles podem
ou não formar um triângulo (cada lado deve ser menor que a soma dos outros dois).

```python
a=float(input("digite o lado a:"))
b=float(input("digite o lado b:"))
c=float(input("digite o lado c:"))
if a<b+c and b<a+c and c<a+b:
  print("forma um triângulo")
else:
  print("não forma um triângulo")
```
